## 6. Hierarchical Clustering

Compute and display a dendrogram using the linkage method `(10 pts)`
- Use `'ward'` linkage to minimize variance within clusters.

```python
linked_data = linkage(ndf_scaled, method='ward')

plt.figure(figsize=(20, 10))
dendrogram(linked_data, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Ward Linkage)')
plt.xlabel('Country Index')
plt.ylabel('Distance')
plt.show()
```

Based on the dendrogram, choose a reasonable number of clusters `(5 pts)`
- Write your chosen number below and explain your reasoning.

 Based on this, a choice of 3 clusters seems reasonable, it intersects three distinct branches. This allows for grouping countries into a few distinct categories without too much fragmentation.

Perform Agglomerative Clustering using your chosen cluster count `(10 pts)`
- Attach the cluster labels back to the original dataframe.

```python
# Perform Agglomerative Clustering with 3 clusters (your chosen number)
agg_cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = agg_cluster.fit_predict(ndf_scaled)

# Add cluster labels to the original dataframe
df['Cluster'] = clusters
df.head()
```
